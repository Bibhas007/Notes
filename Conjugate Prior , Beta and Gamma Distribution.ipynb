# conjugate prior

In Bayesian statistics, a conjugate prior is a specific type of prior distribution that, when combined with a likelihood function, results in a "posterior" distribution that belongs to the same mathematical family as the prior.
Think of it as a mathematical shortcut. It allows you to update your beliefs with new data using simple addition rather than complex calculus

#### The Bayesian Workflow

To understand conjugate priors, you have to look at Bayes' Theorem:

# Posterior∝ Likelihood × Prior

Prior: What you believe before seeing data.(old belief)

Likelihood: What the new data tells you.(new data)

Posterior: Your updated belief after seeing the data.(updated belief)

Usually, multiplying two different types of probability distributions results in a messy, complex formula that is hard to solve. However, if the prior is conjugate to the likelihood, the posterior will have the exact same functional form as the prior.


In Bayes' Theorem, the relationship between these terms is mathematically precise. Here is the breakdown using the standard formula:

##### P(θ∣D)= P(D∣θ)×P(θ) / P(D)


##### 1.The Prior: P(θ)

What it is: Your initial belief about a parameter (θ) before seeing any new data.

Role: It represents existing knowledge, historical data, or even a subjective guess.

Analogy: You believe your friend is usually 10 minutes late to everything.


##### 2.The Likelihood: P(D∣θ)

What it is: The probability of observing the current data (D) given that a specific parameter value (θ) is true.

Role: This is the "voice of the data." It tells you how well the new evidence supports different possible theories.

Analogy: Your friend arrived exactly on time today. How likely is that if they are "normally late" versus "normally punctual"?


##### 3.The Posterior: P(θ∣D)

What it is: Your updated belief about the parameter after accounting for the new data.

Role: This is the final output. It is the result of the "negotiation" between your prior and the likelihood.

Analogy: You now believe your friend is "mostly late, but occasionally punctual." Your belief has shifted.


##### 4.The Evidence (Marginal Likelihood): P(D)

What it is: The total probability of seeing the data under all possible scenarios.

Role: This acts as a normalizing constant. It ensures that the final posterior probabilities sum up to 1 (or 100%). In most conceptual discussions, we ignore this and say the posterior is simply proportional to the top half of the fraction.



#### Example : If you are doing a medical test,

Prior: The general probability of having the disease in the population (e.g., 1%).

Likelihood: How accurate the test is (e.g., the test returns positive 99% of the time if you have it).

Posterior: The actual probability you have the disease given that you just tested positive.


### A Real-World Analogy: The "Blue Paint" Rule
    
Imagine you have a bucket of blue paint (the Prior). Someone hands you a cup of liquid (the Data/Likelihood).

Non-Conjugate: If the liquid is oil, mixing it with your blue paint creates a clumpy, unusable mess. You have to spend hours cleaning it up to understand the new color.

Conjugate: If the liquid is also blue paint (just a different shade), you just pour it in. The result is still "blue paint," just with a slightly different hue. You didn't change the "family" of the substance; you just updated its parameters.

# Beta distribution

The Beta distribution is a family of continuous probability distributions defined on the interval [0, 1].

"probability distribution of probabilities" -- While most distributions model the outcomes of an event (like a coin flip), the Beta distribution models the uncertainty of the likelihood of that event happening.

The Beta distribution is defined by two positive shape parameters, usually denoted as α (alpha) and β (beta). These parameters act as "weights" for two different outcomes:

α (Alpha): Think of this as the count of "successes" plus 1.

β (Beta): Think of this as the count of "failures" plus 1.
    
Real World problem :
The click-through rate of an ad.
The percentage of a battery remaining.
The probability that a drug will be effective.

### The Shape Shifter
	
#### α=1,β=1 -->	  Flat (Uniform)	-->   Every probability is equally likely (we have no data).
#### α>β	  -->       Right-leaning	   -->  The event is more likely to succeed than fail.
#### α<β	  -->       Left-leaning	     -->  The event is more likely to fail than succeed.
#### α=β>1	-->       Bell-shaped	       --> The probability is likely centered around 0.5.
#### α,β<1	 -->      U-shaped	              -->     The outcomes are likely at the extreme ends (0 or 1).

The Beta distribution is the conjugate prior for the Binomial distribution.

You simply add your new successes to α and your new failures to β. It makes updating your beliefs mathematically seamless

#### Beta distribution acts as a conjugate prior for a Binomial likelihood

If our prior belief is Beta(α,β) and we observe k successes and n−k failures in n trials (Binomial likelihood), the posterior is:

Posterior=Beta(α+k,β+(n−k))

Prior: We think the CTR is around 20% but we aren't very sure (Beta(2,8)).

Data: we run an ad and get 40 clicks out of 100 impressions.

Posterior: We update our belief based on this new data.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta, binom

# 1. Define Prior parameters (Initial belief: 2 successes, 8 failures)
alpha_prior, beta_prior = 2, 8

```

```python
# 2. Observed Data (The Likelihood)
n_trials = 100
n_successes = 40
n_failures = n_trials - n_successes

```

```python
# 3. Posterior parameters (The Conjugate Update)
alpha_post = alpha_prior + n_successes
beta_post = beta_prior + n_failures

```

```python
# Setup plotting
x = np.linspace(0, 1, 500)
y_prior = beta.pdf(x, alpha_prior, beta_prior)
y_posterior = beta.pdf(x, alpha_post, beta_post)

```

```python
# Likelihood (scaled to fit on the same plot for visualization)
likelihood = binom.pmf(n_successes, n_trials, x)
likelihood = likelihood / likelihood.max() * y_posterior.max() 

plt.figure(figsize=(10, 6))
plt.plot(x, y_prior, label=f'Prior: Beta({alpha_prior}, {beta_prior})', lw=3, ls='--')
plt.plot(x, y_posterior, label=f'Posterior: Beta({alpha_post}, {beta_post})', lw=3)
plt.fill_between(x, 0, likelihood, alpha=0.1, color='gray', label='Likelihood (Scaled)')

plt.title('Bayesian Updating with Conjugate Priors', fontsize=14)
plt.xlabel('Probability of Success (CTR)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.show()

print(f"Prior Mean: {alpha_prior/(alpha_prior+beta_prior):.2f}")
print(f"Posterior Mean: {alpha_post/(alpha_post+beta_post):.2f}")
```

The Pull of Data: Notice how the Posterior curve (the solid line) shifts toward the Likelihood (the data).

Increased Certainty: The Posterior is much "taller" and "narrower" than the Prior. This represents our increased confidence now that we have seen 100 trials.

The "Weight" of the Prior: If you set your prior α and β to very high numbers (like 200 and 800), the 100 new trials would barely move the curve. This is how you model a very strong pre-existing belief.

#### The Scenario: "The Stubborn Prior"

Imagine you are convinced a coin is fair (Prior: Beta(50,50)). However, you flip it 20 times and it comes up heads every single time (Data: 20 Successes, 0 Failures).

Prior: Deeply believes the probability is 0.5.

Likelihood: Suggests the probability is 1.0.

Posterior: A compromise that settles somewhere in between.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# 1. Strong "Fair Coin" Prior (Total weight = 100 observations)
alpha_prior, beta_prior = 50, 50

# 2. Contradictory Data (20 Heads in a row)
n_successes = 20
n_failures = 0

# 3. Posterior Update
alpha_post = alpha_prior + n_successes
beta_post = beta_prior + n_failures

```

```python
# Plotting
x = np.linspace(0, 1, 500)
plt.figure(figsize=(10, 5))

plt.plot(x, beta.pdf(x, alpha_prior, beta_prior), 'r--', label='Stubborn Prior (Fair Coin)')
plt.plot(x, beta.pdf(x, alpha_post, beta_post), 'b-', label='Updated Posterior')
plt.axvline(1.0, color='green', linestyle=':', label='The Data (100% Success)')

plt.title("When Data Contradicts a Strong Prior")
plt.xlabel("Probability of Success")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"Prior Mean: {alpha_prior/(alpha_prior+beta_prior):.2f}")
print(f"Posterior Mean: {alpha_post/(alpha_post+beta_post):.2f}")
```

The Compromise: The posterior mean shifts to 0.58. Even though the new data showed 100% successes, the prior was so "heavy" (equivalent to 100 previous flips) that the 20 new flips only nudged the needle slightly.

The "Weight" of Evidence: If you want the data to override the prior more easily, you would use a "Weak" or "Uninformative" prior (like Beta(1,1)). In that case, the posterior would jump almost entirely to the right.
    
#### The "Weak Prior" Comparison

If we had used Beta(1,1) as the prior with the same 20 successes:

Posterior: Beta(1+20,1+0)=Beta(21,1)

New Mean: 21/22≈0.95

This illustrates a core Bayesian principle: The more certain you are (the stronger your prior), the more evidence it takes to change your mind.

# Gamma distribution

The Gamma distribution models the waiting time until a certain number of events occur.It is a continuous probability distribution that is strictly positive (from 0 to infinity).

α (Alpha / Shape): The number of events you are waiting for.

β (Beta / Rate): The rate at which events happen.

Note: Sometimes people use θ (Theta / Scale) instead of β, where θ=1/β. In that case, θ represents the average time between events.

#### Real time problem 
Insurance: How long until the next 10 claims are filed?
Reliability: How long until a machine part fails for the 3rd time?
Meteorology: The total amount of rainfall in a season (modeled as a series of storm events).
    
#### Relationship to Other Distributions
The Gamma distribution is like the "parent" of several other famous distributions:

Exponential Distribution: If you set α=1, you get the Exponential distribution. This models the time until the very first event happens.

Chi-Squared Distribution: If you set α=ν/2 and β=1/2, you get the χ 
2
  distribution with ν degrees of freedom.

Poisson Distribution: If the Poisson distribution tells you how many events happen in 1 hour, the Gamma distribution tells you how many hours it takes for k events to happen.

#### Gamma distribution is the conjugate prior for the Poisson distribution.
#### The Components
Likelihood (Poisson): Models the number of events k occurring in a fixed interval, given a rate λ (lambda).

Prior (Gamma): Models our initial belief about what that rate λ is. It has two parameters:

α (Shape): Can be thought of as "total prior events observed."

β (Rate): Can be thought of as "total prior time intervals elapsed."

If your prior is Gamma(α,β) and you observe k new events over t new time intervals, your posterior is:

Posterior=Gamma(α+k,β+t)

Suppose you think a coffee shop gets 10 customers per hour on average (α=10,β=1). Then, you sit in the shop for 3 hours and count only 12 customers total.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, poisson

# 1. Prior: We think 10 customers/hour (alpha=10, beta=1)
alpha_prior = 10
beta_prior = 1 # The rate parameter (inverse scale)

# 2. Data: 12 customers observed over 3 hours
k_observed = 12
t_observed = 3

# 3. Posterior Update
alpha_post = alpha_prior + k_observed
beta_post = beta_prior + t_observed

```

```python
# Plotting
x = np.linspace(0, 20, 500)
plt.figure(figsize=(10, 5))

plt.plot(x, gamma.pdf(x, alpha_prior, scale=1/beta_prior), 'r--', label='Prior (Exp. 10/hr)')
plt.plot(x, gamma.pdf(x, alpha_post, scale=1/beta_post), 'b-', label='Posterior (Updated)')

plt.title("Bayesian Update: Poisson Rate (λ)")
plt.xlabel("Arrival Rate (Customers per hour)")
plt.ylabel("Density")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

print(f"Prior Mean Rate: {alpha_prior/beta_prior:.2f} per hour")
print(f"Posterior Mean Rate: {alpha_post/beta_post:.2f} per hour")
```

Small Data Problems: If you only watch the shop for 1 hour and see 0 people, a Frequentist might say "the arrival rate is 0." A Bayesian using a Gamma prior would say "the rate has dropped, but I still believe people will eventually show up."

Reliability Engineering: Gamma-Poisson is used to predict how many times a machine will fail over the next year based on historical data.

#### Problem 1: The Website Bug Tracker
    
Scenario: You believe a software module has an average of 2 bugs per week (α=2,β=1). Over the next 4 weeks, you find a total of 15 bugs.

Task: Calculate the posterior mean and plot the change.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Prior: 2 bugs / 1 week
a_prior, b_prior = 2, 1

# Data: 15 bugs / 4 weeks
bugs_found = 15
weeks_passed = 4

# Posterior Update
a_post = a_prior + bugs_found
b_post = b_prior + weeks_passed
```

```python

x = np.linspace(0, 10, 500)
plt.plot(x, gamma.pdf(x, a_prior, scale=1/b_prior), label='Prior (2/week)', ls='--')
plt.plot(x, gamma.pdf(x, a_post, scale=1/b_post), label='Posterior (Updated)')
plt.title("Bug Rate Estimation")
plt.xlabel("Bugs per Week")
plt.legend()
plt.show()

print(f"New expected bug rate: {a_post/b_post:.2f} per week")
```

#### Problem 2: Call Center Volume (Weak Prior)

Scenario: You have no idea how many calls a new center gets per hour, so you use a "Flat" or "Weak" prior: Gamma(1,0.001). In the first hour, they receive 50 calls.

Task: Use Python to find the 90% Credible Interval for the call rate.

```python
from scipy.stats import gamma

# Weak Prior
a_prior, b_prior = 1, 0.001

# Data
calls = 50
hours = 1

# Update
a_post = a_prior + calls
b_post = b_prior + hours

# 90% Credible Interval (5th to 95th percentile)
ci_low = gamma.ppf(0.05, a_post, scale=1/b_post)
ci_high = gamma.ppf(0.95, a_post, scale=1/b_post)

print(f"90% Credible Interval for Calls/Hour: [{ci_low:.2f}, {ci_high:.2f}]")
```

#### Problem 3: Comparing Two Servers (A/B Testing Rates)

Scenario: Server A has a prior of Gamma(10,2). Server B has a prior of Gamma(10,2). In a test, Server A handles 30 requests in 5 minutes. Server B handles 45 requests in 5 minutes.

Task: Calculate the posterior means and determine which server is likely faster.

```python
# Server A
a_post_A = 10 + 30
b_post_A = 2 + 5  # assuming units are minutes

# Server B
a_post_B = 10 + 45
b_post_B = 2 + 5

mean_A = a_post_A / b_post_A
mean_B = a_post_B / b_post_B

print(f"Server A Rate: {mean_A:.2f} req/min")
print(f"Server B Rate: {mean_B:.2f} req/min")
print(f"Difference: {mean_B - mean_A:.2f} req/min")
```

#### Problem 4: Probability of Exceeding a Threshold

Scenario: Using the posterior from Problem 1 (Gamma(17,5)), what is the probability that the bug rate is actually higher than 4 bugs per week?

```python
# Probability of rate > 4 is 1 - CDF(4)
prob_high_rate = 1 - gamma.cdf(4, 17, scale=1/5)

print(f"Probability that the bug rate is > 4: {prob_high_rate:.4f}")
```

#### Problem 5: Visualizing "Information Gain"

Scenario: Show how the Gamma distribution narrows (becomes more certain) as we increase the observation time. Compare t=1,t=10, and t=100 while keeping the average rate at 5.

```python
x = np.linspace(0, 10, 1000)
times = [1, 10, 100]
rate = 5

for t in times:
    events = rate * t
    # Using a weak prior (1, 0.1)
    a = 1 + events
    b = 0.1 + t
    plt.plot(x, gamma.pdf(x, a, scale=1/b), label=f'Observed for {t} units')

plt.title("Certainty Increases with Observation Time")
plt.legend()
plt.show()
```
